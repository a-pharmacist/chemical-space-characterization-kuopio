# Starting the environment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
s_aureus_data = pd.read_csv("/content/s_aureus_43300_descriptors_and_lipinski.csv")
bracoli_data = pd.read_csv("/content/bracoli_descriptors.csv")
nubbe_data = pd.read_csv("/content/nubbe_descriptors.csv")
```

```python
molecular_properties = [col for col in s_aureus_data.columns[3:-1]]
molecular_properties
```

# Molecular weight

```python
column = "MolWt"

sns.set(style="whitegrid", palette="Blues")

all_data = pd.concat([s_aureus_data[column], bracoli_data[column], nubbe_data[column]])
x_min, x_max = all_data.min(), all_data.max()
bins = np.linspace(x_min, x_max, 30)

fig, axs = plt.subplots(1, 3, figsize=(14, 5), sharey=True)

datasets = [s_aureus_data[column], bracoli_data[column], nubbe_data[column]]
titles = ["S. aureus", "BraCoLi", "NuBBE"]
colors = [sns.color_palette("Blues")[3], sns.color_palette("Blues")[4], sns.color_palette("Blues")[5]]

for ax, data, title, color in zip(axs, datasets, titles, colors):
    counts, bin_edges, patches = ax.hist(data, bins=bins, color=color, edgecolor='black', linewidth=0.3)
    ax.set_title(title)
    ax.set_xlim(x_min, x_max)
    ax.set_xlabel("")
    ax.set_ylabel("")
    for count, x in zip(counts, bin_edges[:-1]):
        if count > 0:
            ax.text(x + (bin_edges[1] - bin_edges[0]) / 2, count + 0.5, str(int(count)),
                    ha='center', va='bottom', fontsize=8)
    sns.lineplot(x=(bin_edges[:-1] + bin_edges[1:]) / 2, y=counts, ax=ax, color='black', linewidth=1, alpha=0.5)

for ax in axs:
    sns.despine(ax=ax)

fig.suptitle(f"{column} Distribution Along Libraries", fontsize=16, fontweight='bold')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
